### Cleaning and constructing Key Measurables

Injuries and robustness to destruction are a key interest point for the client. Clean and impute relevant columns and then create derived fields that best quantifies what the client wishes to track. **Use commenting or markdown to explain any cleaning assumptions as well as any derived columns you create.**

```python
injury_cols = [
    'total_fatal_injuries',
    'total_serious_injuries',
    'total_minor_injuries',
    'total_uninjured'
]

# Fill missing values
df3.loc[:, injury_cols] = df3.loc[:, injury_cols].fillna(0)

# Replace negatives with 0
df3.loc[:, injury_cols] = df3.loc[:, injury_cols].clip(lower=0)

# Convert to integer
df3.loc[:, injury_cols] = df3.loc[:, injury_cols].astype(int)
```

```python
# Convert to num
for col in injury_cols:
    df3[col] = pd.to_numeric(df3[col], errors='coerce')
```

```python
#Total passengers
df3['total_passengers'] = (
    df3['total_fatal_injuries'] +
    df3['total_serious_injuries'] +
    df3['total_minor_injuries'] +
    df3['total_uninjured']
)
```

```python
# Serious + Fatal injuries
df3['serious_fatal'] = (
    df3['total_fatal_injuries'] +
    df3['total_serious_injuries']
)
```

```python
# Injury fraction
df3['injury_fraction'] = df3['serious_fatal'] / df3['total_passengers']
```

**Construct metric for fatal/serious injuries**

*Hint:* Estimate the total number of passengers on each flight. The likelihood of serious / fatal injury can be estimated as a fraction from this.

```python
df3['total_aboard_est'] = df3.loc[:, injury_cols].sum(axis=1)
```

```python
df3 = df3.loc[df3['total_aboard_est'] > 0]
```

```python
# Fatal injury rate
df3['fatal_injury_rate'] = (
    df3['total_fatal_injuries'] / df3['total_aboard_est']
)
```

```python
# Serious injury rate
df3['serious_injury_rate'] = (
    df3['total_serious_injuries'] / df3['total_aboard_est']
)
```

```python
# survival rate
df3['survival_rate'] = (
    df3['total_uninjured'] / df3['total_aboard_est']
)
```

```python
df3['safety_score'] = df3['survival_rate'] - (
    df3['fatal_injury_rate'] + df3['serious_injury_rate']
)
```

```python

# handle missing values using loc
df3.loc[df3['aircraft_damage'].isnull(), 'aircraft_damage'] = 'unknown'

# remove unwanted characters
df3['aircraft_damage'] = df3['aircraft_damage'].str.replace(
    r'[^a-zA-Z\s]', '', regex=True
)
```

### Investigate the *Make* column
- Identify cleaning tasks here
- List cleaning tasks clearly in markdown
- Execute the cleaning tasks
- For your analysis, keep Makes with a reasonable number (you can put the threshold at 50 though lower could work as well)

```python
# handle missing values first
df3.loc[df3['make'].isna(), 'make'] = 'unknown'

# standardize text
df3['make'] = df3['make'].str.lower().str.strip()

# remove unwanted characters
df3['make'] = df3['make'].str.replace(r'[^a-z\s]', '', regex=True)

# fix empty strings
df3.loc[df3['make'] == '', 'make'] = 'unknown'

# group rare makes
make_counts = df3['make'].value_counts()
valid_makes = make_counts[make_counts >= 50].index

# create cleaned column using loc
df3['make_cleaned'] = df3['make']
df3.loc[~df3['make'].isin(valid_makes), 'make_cleaned'] = 'other'
```
